# Q1) The Probability of Shooting at a Target

A point is chosen at random from a disk of radius $10$. Let $A$ be the event that the point lies within $1$ unit of the boundary.

1. Model the experiment assuming the point is uniformly distributed with respect to area, and compute $P(A)$.
2. Model the experiment assuming the distance to the point from the center is chosen uniformly from $[0,10]$, and the direction is chosen independently and uniformly from $[0,2\pi)$ and compute $P(A)$.
3. Explain why the two answers differ.


Let the disk be

$$
\Omega=\{(x,y)\in \mathbb{R}^2 : x^2+y^2\le 10^2\}.
$$

Let $A$ be the event that the chosen point lies within $1$ unit of the boundary. Since the boundary is the circle of radius $10$,

$$
A=\left\{(x,y)\in \Omega : 9\le \sqrt{x^2+y^2}\le 10\right\}.
$$

Thus, $A$ is the outer annulus of the disk.

We consider two probability models.

---

**Uniform distribution with respect to area**

In this model, probability is proportional to area.

The total area of the disk is

$$
\pi(10)^2=100\pi.
$$

The area of the annulus is

$$
\pi(10)^2-\pi(9)^2
=100\pi-81\pi
=19\pi.
$$

Therefore,

$$
P(A)=\frac{19\pi}{100\pi}
=\frac{19}{100}.
$$

Hence,

$$
\boxed{P(A)=\frac{19}{100}}.
$$

---

**Uniform distribution of the radius**

In this model, the radius is chosen uniformly from $[0,10]$, and the angle is chosen independently from $[0,2\pi)$.

The event $A$ corresponds to

$$
9\le r\le 10.
$$

Thus,

$$
P(A)=\frac{10-9}{10-0}
=\frac{1}{10}.
$$

Hence,

$$
\boxed{P(A)=\frac{1}{10}}.
$$

---

**Why the two answers differ**

The phrase “chosen at random” does not determine a unique probability model.

In the first model, probability is distributed uniformly over area, giving

$$
P(A)=\frac{19}{100}.
$$

In the second model, probability is distributed uniformly over radius, giving

$$
P(A)=\frac{1}{10}.
$$

The event $A$ is the same in both cases, but the probability measure is different.

---

**Probability of Throwing Darts**

If darts are thrown randomly at a circular board without directional or radial bias, the probability of landing in a region is proportional to its area.

Thus, the probability of landing within $1$ unit of the boundary is

$$
\frac{19}{100}.
$$

The radius-uniform model represents a different sampling mechanism: first choosing a radius uniformly and then choosing a direction.

So:

* **dart throwing** $\to$ area-uniform,
* **two-step radius selection** $\to$ radius-uniform.

In practical situations, “randomly chosen point in a disk” is usually modeled by the uniform distribution on area.

# Q2) Bertrand's Paradox

A chord of a circle is chosen “at random.” What is the probability that its length exceeds the side length of the inscribed equilateral triangle?

Show that different mathematically natural models of “random chord” lead to different answers.

**Preliminary geometric fact**

Let the circle be

$$
x^2+y^2=1.
$$

The side length of the inscribed equilateral triangle is

$$
\sqrt{3}.
$$

A chord is longer than $\sqrt{3}$ if and only if its perpendicular distance from the center is less than $\dfrac{1}{2}$.

Thus, if a chord is at distance $r\in[0,1]$ from the center, its length is

$$
L(r)=2\sqrt{1-r^2}.
$$

The condition

$$
L(r)>\sqrt{3}
$$

gives

$$
r<\frac12.
$$

Hence, the event of interest is

$$
A=\{\text{distance of chord from center}<1/2\}.
$$

This model can be described as:

* choose one endpoint uniformly on the circumference,
* choose the other endpoint uniformly on the circumference.

By rotational symmetry, the first endpoint may be fixed and the second chosen uniformly.

---

**Probability space**

Let

$$
\Omega_1=[0,\pi]
$$

with Borel sigma-algebra

$$
\mathfrak A_1=\mathfrak B([0,\pi]),
$$

and probability measure

$$
P_1(E)=\frac{\lambda(E)}{\pi},
\qquad E\in\mathfrak A_1,
$$

where $\lambda$ is Lebesgue measure.

Here, $\theta\in[0,\pi]$ represents the smaller central angle subtended by the chord.

---

**Event**

A chord subtending angle $\theta$ has length

$$
L(\theta)=2\sin(\theta/2).
$$

The condition

$$
L(\theta)>\sqrt{3}
$$

gives

$$
2\sin(\theta/2)>\sqrt{3}
$$

and hence

$$
\sin(\theta/2)>\frac{\sqrt{3}}{2}.
$$

Thus,

$$
\theta>2\pi/3.
$$

So,

$$
A_1=(2\pi/3,\pi].
$$

---

**Probability**

Therefore,

$$
P_1(A_1)=\frac{\pi-2\pi/3}{\pi}
=\frac{1}{3}.
$$

Hence, under Model-1,

$$
\boxed{P_1(A_1)=\frac13}.
$$

This model is:

* choose a radius,
* choose a point uniformly along that radius,
* draw the chord perpendicular to the radius through that point.

Equivalently, the distance of the chord from the center is chosen uniformly in $[0,1]$.

---

**Probability space**

Let

$$
\Omega_2=[0,1]
$$

with sigma-algebra

$$
\mathfrak A_2=\mathfrak B([0,1]),
$$

and probability measure

$$
P_2(E)=\lambda(E),
\qquad E\in\mathfrak A_2,
$$

where $\lambda([0,1])=1$.

Here, $r\in[0,1]$ represents the perpendicular distance from the center to the chord.

---

**Event**

The chord is longer than the side of the inscribed equilateral triangle when

$$
r<\frac12.
$$

Thus,

$$
A_2=[0,1/2).
$$

---

**Probability**

Therefore,

$$
P_2(A_2)=\lambda([0,1/2))
=\frac12.
$$

Hence, under Model 2,

$$
\boxed{P_2(A_2)=\frac12}.
$$

This model is:

* choose the midpoint of the chord uniformly from the whole disk,
* draw the unique chord having that midpoint.

Since every chord is uniquely determined by its midpoint, this defines a probability law on chords.

---

**Probability space**

Let

$$
\Omega_3=\{(x,y)\in\mathbb R^2:x^2+y^2\le 1\}
$$

be the closed unit disk, with sigma-algebra

$$
\mathfrak A_3=\mathfrak B(\Omega_3).
$$

Define $P_3$ by normalized area:

$$
P_3(E)=\frac{\lambda_2(E)}{\pi},
\qquad E\in\mathfrak A_3,
$$

where $\lambda_2$ is planar Lebesgue measure.

Here, $(x,y)$ represents the midpoint of the chord.

---

**Event**

If the midpoint is $(x,y)$, its distance from the center is

$$
r=\sqrt{x^2+y^2}.
$$

The chord is longer than $\sqrt{3}$ when

$$
r<\frac12.
$$

Thus,

$$
A_3=\{(x,y)\in\Omega_3:x^2+y^2<1/4\}.
$$

This is the disk of radius $\dfrac12$.

---

**Probability**

Therefore,

$$
P_3(A_3)=\frac{\text{area of disk of radius }1/2}{\text{area of unit disk}}
=\frac{\pi(1/2)^2}{\pi}
=\frac14.
$$

Hence, under Model 3,

$$
\boxed{P_3(A_3)=\frac14}.
$$

We obtain three different answers:

| Model | Parameter chosen uniformly | Probability |
|-------|----------------------------|-------------|
| 1 | endpoint angle / subtended angle | $\frac{1}{3}$ |
| 2 | distance from center | $\frac{1}{2}$ |
| 3 | midpoint in the disk | $\frac{1}{4}$ |

Each answer is correct under a different mathematical interpretation of “choose a chord at random.”

This is Bertrand’s paradox: the verbal description alone does not determine a unique probability model. The sampling procedure, or equivalently the probability space, must be specified.

---

Let the circle have radius $1$. A chord exceeds the side length of the inscribed equilateral triangle if and only if its distance from the center is less than $\dfrac{1}{2}$.

Three natural models of a “random chord” are:

1. **Uniform endpoint model:**  
   Choose one endpoint uniformly and then the other uniformly.  
   The central angle $\theta\in[0,\pi]$ is uniform, and the chord is longer when

   $$
   \theta>2\pi/3.
   $$

   Therefore,

   $$
   P=\frac{\pi-2\pi/3}{\pi}=\frac13.
   $$

2. **Uniform distance model:**  
   Choose the perpendicular distance $r$ from the center uniformly in $[0,1]$.  
   The chord is longer when

   $$
   r<\frac12.
   $$

   Therefore,

   $$
   P=\frac12.
   $$

3. **Uniform midpoint model:**  
   Choose the midpoint uniformly in the unit disk.  
   The chord is longer when the midpoint lies in the disk of radius $\dfrac12$.

   Therefore,

   $$
   P=\frac{\pi(1/2)^2}{\pi}=\frac14.
   $$

Thus, the phrase “a chord is chosen at random” is ambiguous, since different probability models produce different answers.

# Q9) Information Available in a Discrete Random Variable

A factory inspects a product and records its quality using two different coding systems.

Each product falls into exactly one of the following four categories:

\Omega=\{G_1,G_2,D_1,D_2\},

$$
\Omega=\{G_1,G_2,D_1,D_2\},
$$

$$
\mathfrak F=\mathfrak P(\Omega),
$$

and the probability measure $P$ is defined by

$$
P(\{G_1\})=0.50,\qquad P(\{G_2\})=0.20,\qquad P(\{D_1\})=0.10,\qquad P(\{D_2\})=0.20.
$$

Since

$$
0.50+0.20+0.10+0.20=1,
$$

this is a valid probability measure.

---

2. The random variable $X$ is defined by

$$
X(G_1)=0,\qquad X(G_2)=0,\qquad X(D_1)=1,\qquad X(D_2)=1.
$$

Thus,

$$
X^{-1}(\{0\})=\{G_1,G_2\},
$$

and

$$
X^{-1}(\{1\})=\{D_1,D_2\}.
$$

Hence,

$$
\sigma(X)=\left\{\varnothing,\Omega,\{G_1,G_2\},\{D_1,D_2\}\right\}.

---

3. The random variable $Y$ is defined by

$$
Y(G_1)=1,\qquad Y(G_2)=2,\qquad Y(D_1)=3,\qquad Y(D_2)=4.

$$
\sigma(Y)=\mathfrak P(\Omega).

Explicitly,

$$
\sigma(Y)=\left\{\varnothing,\{G_1\},\{G_2\},\{D_1\},\{D_2\},\{G_1,G_2\},\{G_1,D_1\},\ldots,\Omega\right\}.

---

4. Since $\sigma(X)\subseteq \sigma(Y)$, the variable $Y$ contains more information than $X$.

5. The distribution of $X$ is

---

9. Conditioning on $Y$ provides more detailed information than conditioning on $X$. Knowing $Y$ determines the exact outcome, while knowing $X$ only distinguishes between good and defective items.

10. The $\sigma$-algebra $\sigma(X)$ only distinguishes quality classes, while $\sigma(Y)=\mathfrak P(\Omega)$ distinguishes every outcome. Therefore,

so $Y$ carries more information than $X$.

Using natural logarithms, the Shannon entropy is

$$
H(Z)=-\sum_z P(z)\ln P(z).
$$

---

$$
H(X)=-(0.70\ln 0.70+0.30\ln 0.30),
$$

$$
H(Y)=-(0.50\ln 0.50+0.20\ln 0.20+0.10\ln 0.10+0.20\ln 0.20).
$$

Thus,

$$
H(X)\approx 0.61086,
$$

$$
H(Y)\approx 1.22061.
$$

---

For the joint entropy,

$$
H(X,Y)= -\sum_{x,y} P(x,y)\ln P(x,y),
$$

so only the nonzero terms contribute:

$$
H(X,Y)=-(0.50\ln 0.50+0.20\ln 0.20+0.10\ln 0.10+0.20\ln 0.20).
$$

Therefore,

$$
H(X,Y)\approx 1.22061.
$$

---

Since each value of $Y$ determines $X$ exactly,

$$
H(X\mid Y)=0.
$$

Also,

$$
H(Y\mid X)=H(X,Y)-H(X),
$$

so

$$
H(Y\mid X)\approx 1.22061-0.61086=0.60975.
$$

---

Final values:

$$
H(X)\approx 0.61086,
$$

$$
H(Y)\approx 1.22061,
$$

$$
H(X,Y)\approx 1.22061,
$$

$$
H(X\mid Y)=0,
$$

$$
H(Y\mid X)\approx 0.60975.
$$

---

Interpretation:

$H(X)$ measures uncertainty about whether an item is good or defective.

$H(Y)$ measures uncertainty about the full classification.

$H(X\mid Y)=0$ because $Y$ determines $X$ exactly.

$H(Y\mid X)>0$ because knowing only $X$ does not determine the machine.

Thus, entropy confirms that $Y$ contains more information than $X$.

1. The probability space is

$$
\Omega=\mathbb R^2, \qquad \mathfrak F=\mathfrak B(\mathbb R^2),
$$

where $\mathfrak B(\mathbb R^2)$ is the Borel $\sigma$-algebra on $\mathbb R^2$.

A point $\omega\in\Omega$ has the form

$$
\omega=(t,p),
$$

where $t$ is the standardized temperature deviation and $p$ is the standardized pressure deviation.

The probability measure $P$ is defined by the joint density

$$
f_{T,P}(t,p)=\frac{1}{4\pi}\exp\left(-\frac{t^2}{2}-\frac{p^2}{8}\right), \qquad (t,p)\in\mathbb R^2.
$$

Thus for any Borel set $A\subseteq\mathbb R^2$,

$$
P(A)=\iint_A f_{T,P}(t,p)\,dt\,dp.
$$

Since $T\sim N(0,1)$ and $P\sim N(0,4)$ are independent, this defines a valid probability measure.

---

2. The random variable $X$ is defined by

$$
X(t,p)=t.
$$

Thus $X$ records only the temperature deviation.

The $\sigma$-algebra generated by $X$ is

$$
\sigma(X)=\{X^{-1}(B): B\in\mathfrak B(\mathbb R)\}.
$$

Since

$$
X^{-1}(B)=\{(t,p)\in\mathbb R^2:t\in B\}=B\times \mathbb R,

we obtain

$$
\sigma(X)=\{B\times\mathbb R: B\in\mathfrak B(\mathbb R)\}.
$$

---

3. The random variable $Y$ is defined by

$$
Y(t,p)=(t,p).
$$

Thus $Y$ records both temperature and pressure.

Since $Y$ is the identity map,

$$
\sigma(Y)=\mathfrak B(\mathbb R^2).
$$

---

4. We have

$$
\sigma(X)\subset \sigma(Y),
$$

and the inclusion is strict since sets depending on both coordinates lie in $\sigma(Y)$ but not in $\sigma(X)$.

---

5. The marginal density of $X$ is

$$
f_X(x)=\int_{-\infty}^{\infty} f_{T,P}(x,p)\,dp
=\frac{1}{\sqrt{2\pi}}e^{-x^2/2}.
$$

Hence

$$
X\sim N(0,1).
$$

---

6. The marginal density of $Y$ is

$$
f_Y(t,p)=\frac{1}{4\pi}\exp\left(-\frac{t^2}{2}-\frac{p^2}{8}\right).
$$

---

7. Basic probabilities:

$$
P(X\le 0)=\frac12,
\qquad
P(X>1)=1-\Phi(1)\approx 0.1587.
$$

Also, by independence,

$$
P(-1\le T\le 1)=2\Phi(1)-1,
\qquad
P(-2\le P\le 2)=2\Phi(1)-1.
$$

Thus

$$
P([-1,1]\times[-2,2])=(2\Phi(1)-1)^2\approx 0.466.
$$

---

8. Since $Y$ determines $X$,

$$
P(X=t\mid Y=(t,p))=1,
\qquad
P_{X|Y=(t,p)}=\delta_t.
$$

---

9. The conditional distribution of $Y$ given $X=t$ is

$$
Y|X=t \sim (t,P), \quad P\sim N(0,4).
$$

---

10. Conditioning on $X$ leaves uncertainty in $P$, while conditioning on $Y$ removes all uncertainty about both variables.

---

11. Define

$$
\pi_1(t,p)=t.
$$

Then

$$
X=\pi_1\circ Y,
\qquad
\sigma(X)\subseteq\sigma(Y).
$$

---

12. Visualization (conceptual):

- $f_Y(t,p)$ is a bivariate normal density,
- $f_X(x)$ is a standard normal density,
- $\sigma(X)$ corresponds to vertical strips $B\times\mathbb R$,
- $\sigma(Y)$ corresponds to all Borel subsets of $\mathbb R^2$.

---

13. Differential entropy of $X$:

$$
h(X)=\frac12\ln(2\pi e)\approx 1.4189.
$$

---

14. Differential entropy of $Y$:

$$
h(Y)=h(T)+h(P)
=\frac12\ln(2\pi e)+\frac12\ln(8\pi e)
=\ln(2\pi e)+\ln 2
\approx 3.5310.
$$

---

15. Conditional entropy:

$$
h(Y|X)=h(P)=\frac12\ln(8\pi e)\approx 2.1121.
$$

---

16. Since $X$ is a deterministic function of $Y$, in the continuous case

$$
h(X|Y) \to -\infty,
$$

showing the difference between discrete and differential entropy.

---

**Conclusion**

$$
\sigma(X)\subseteq\sigma(Y),
$$

and $Y$ contains strictly more information than $X$. This is reflected in both measure-theoretic structure and entropy-based analysis.

$$

````python
The following code provides a visualization of the two $\sigma$-algebras in $\mathbb R^2$ for the example

$$
X(t,p)=t, \qquad Y(t,p)=(t,p).
$$

---

The three panels represent:

1. an event in $\sigma(X)$,
2. an event in $\sigma(Y)$,
3. a direct comparison showing the inclusion relation.

---

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle

fig, axes = plt.subplots(1, 3, figsize=(15, 4.8))

xlim = (-4, 4)
ylim = (-8, 8)

# -------------------------
# (1) sigma(X)
# -------------------------
ax = axes[0]
ax.set_title(r"Event in $\sigma(X)$: $B\times\mathbb{R}$")
ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.set_xlabel(r"$t$")
ax.set_ylabel(r"$p$")

strip = Rectangle((-1, ylim[0]), 2, ylim[1] - ylim[0],
                  facecolor="skyblue", edgecolor="navy", alpha=0.55)
ax.add_patch(strip)

ax.text(0, 0, r"$[-1,1]\times \mathbb{R}$",
        ha="center", va="center", fontsize=12,
        bbox=dict(facecolor="white", edgecolor="none", alpha=0.8))

ax.axhline(0, color="black", lw=0.8)
ax.axvline(0, color="black", lw=0.8)
ax.grid(alpha=0.25)

# -------------------------
# (2) sigma(Y)
# -------------------------
ax = axes[1]
ax.set_title(r"Event in $\sigma(Y)$: general Borel subset")
ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.set_xlabel(r"$t$")
ax.set_ylabel(r"$p$")

rect = Rectangle((-1.5, -3), 3, 6,
                 facecolor="lightgreen", edgecolor="darkgreen", alpha=0.55)
ax.add_patch(rect)

disk = Circle((2.2, 3.5), 1.2,
              facecolor="salmon", edgecolor="darkred", alpha=0.45)
ax.add_patch(disk)

ax.text(0, 0, r"$[-1.5,1.5]\times[-3,3]$",
        ha="center", va="center", fontsize=11,
        bbox=dict(facecolor="white", edgecolor="none", alpha=0.8))

ax.text(2.2, 3.5, r"disk",
        ha="center", va="center", fontsize=10,
        bbox=dict(facecolor="white", edgecolor="none", alpha=0.8))

ax.axhline(0, color="black", lw=0.8)
ax.axvline(0, color="black", lw=0.8)
ax.grid(alpha=0.25)

# -------------------------
# (3) comparison
# -------------------------
ax = axes[2]
ax.set_title(r"Comparison: $\sigma(X)\subseteq\sigma(Y)$")
ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.set_xlabel(r"$t$")
ax.set_ylabel(r"$p$")

strip2 = Rectangle((-1, ylim[0]), 2, ylim[1] - ylim[0],
                   facecolor="skyblue", edgecolor="navy", alpha=0.35)
ax.add_patch(strip2)

rect2 = Rectangle((-1.5, -3), 3, 6,
                  facecolor="lightgreen", edgecolor="darkgreen", alpha=0.5)
ax.add_patch(rect2)

ax.text(0, 6.5, r"$\sigma(X)$: vertical strips only",
        ha="center", fontsize=11,
        bbox=dict(facecolor="white", edgecolor="none", alpha=0.8))

ax.text(0, -6.5, r"$\sigma(Y)$: full Borel subsets",
        ha="center", fontsize=11,
        bbox=dict(facecolor="white", edgecolor="none", alpha=0.8))

ax.axhline(0, color="black", lw=0.8)
ax.axvline(0, color="black", lw=0.8)
ax.grid(alpha=0.25)

plt.tight_layout()
plt.show()
````

````python
The following code visualizes the probability structure of the example

$$
X(t,p)=t, \qquad Y(t,p)=(t,p).
$$

It illustrates:

* the joint density of $Y=(T,P)$,
* the marginal densities of $X$ and $P$,
* the conditional distributions $P\mid X$ and $X\mid Y$.

---

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu_T, sigma_T = 0.0, 1.0
mu_P, sigma_P = 0.0, 2.0

# Grids
t = np.linspace(-4, 4, 400)
p = np.linspace(-8, 8, 400)
T, P = np.meshgrid(t, p)

# Joint density of Y = (T, P)
joint_pdf = (
    norm.pdf(T, loc=mu_T, scale=sigma_T) *
    norm.pdf(P, loc=mu_P, scale=sigma_P)
)

# Marginal density of X = T
fx = norm.pdf(t, loc=mu_T, scale=sigma_T)

# Marginal density of P
fp = norm.pdf(p, loc=mu_P, scale=sigma_P)

# Conditional density P | X = t0
t0 = 1.0
cond_p_given_x = norm.pdf(p, loc=mu_P, scale=sigma_P)

# Conditional X | Y = (t1, p1) (degenerate, approximated)
t1, p1 = 1.0, 2.0
eps = 0.08
cond_x_given_y_vis = norm.pdf(t, loc=t1, scale=eps)

fig = plt.figure(figsize=(12, 12))
gs = fig.add_gridspec(3, 2, height_ratios=[1.1, 1, 1], hspace=0.45, wspace=0.28)

# (1) Joint density
ax1 = fig.add_subplot(gs[0, :])
cs = ax1.contourf(T, P, joint_pdf, levels=25, cmap="viridis")
ax1.contour(T, P, joint_pdf, levels=10, colors="white", linewidths=0.5, alpha=0.6)
ax1.set_title(r"Joint density of $Y=(T,P)$")
ax1.set_xlabel(r"$t$")
ax1.set_ylabel(r"$p$")
fig.colorbar(cs, ax=ax1, shrink=0.9)

# (3) Marginal of X
ax3 = fig.add_subplot(gs[1, 0])
ax3.plot(t, fx, color="navy", lw=2)
ax3.fill_between(t, fx, alpha=0.35)
ax3.set_title(r"Marginal density of $X=T$")
ax3.set_xlabel(r"$x$")
ax3.set_ylabel(r"$f_X(x)$")
ax3.grid(alpha=0.25)

# (4) Marginal of P
ax4 = fig.add_subplot(gs[1, 1])
ax4.plot(p, fp, color="darkgreen", lw=2)
ax4.fill_between(p, fp, alpha=0.35)
ax4.set_title(r"Marginal density of $P$")
ax4.set_xlabel(r"$p$")
ax4.set_ylabel(r"$f_P(p)$")
ax4.grid(alpha=0.25)

# (5) Conditional P | X
ax5 = fig.add_subplot(gs[2, 0])
ax5.plot(p, cond_p_given_x, color="crimson", lw=2)
ax5.fill_between(p, cond_p_given_x, alpha=0.35)
ax5.set_title(r"Conditional density $P \mid X=t_0$")
ax5.set_xlabel(r"$p$")
ax5.set_ylabel(r"$f_{P|X}(p)$")
ax5.grid(alpha=0.25)

# (6) Conditional X | Y
ax6 = fig.add_subplot(gs[2, 1])
ax6.plot(t, cond_x_given_y_vis, color="purple", lw=2)
ax6.axvline(t1, color="black", ls="--", lw=1.3)
ax6.set_title(r"$X \mid Y=(t_1,p_1)$ (degenerate at $t_1$)")
ax6.set_xlabel(r"$x$")
ax6.set_ylabel("Density (approx.)")
ax6.grid(alpha=0.25)

plt.show()
````
