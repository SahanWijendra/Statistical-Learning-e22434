A bank monitors credit-card transactions for possible fraud.

For each transaction, let the sample space be

$$
\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\},
$$

where:

* $\omega_1$: low-risk customer, legitimate transaction,
* $\omega_2$: low-risk customer, fraudulent transaction,
* $\omega_3$: high-risk customer, legitimate transaction,
* $\omega_4$: high-risk customer, fraudulent transaction.

Assume the probabilities are

$$
P(\{\omega_1\})=0.50,\qquad
P(\{\omega_2\})=0.10,\qquad
P(\{\omega_3\})=0.20,\qquad
P(\{\omega_4\})=0.20.
$$

Let $X:\Omega\to\mathbb R$ denote the fraud indicator:

$$
X(\omega_1)=0,\qquad
X(\omega_2)=1,\qquad
X(\omega_3)=0,\qquad
X(\omega_4)=1.
$$

Thus $X=1$ means the transaction is fraudulent, and $X=0$ means it is legitimate.

Now suppose the bank does not observe the full outcome $\omega$, but only the customer's risk class. Define $Y:\Omega\to\mathbb R$ by

$$
Y(\omega_1)=0,\qquad
Y(\omega_2)=0,\qquad
Y(\omega_3)=1,\qquad
Y(\omega_4)=1.
$$

Thus:

* $Y=0$ means “low-risk customer,”
* $Y=1$ means “high-risk customer.”

We compute the Minimum Mean Square Error (MMSE), given by

$$
\text{MSE}_{\min} = \mathbb{E}\big[(X - \mathbb{E}[X \mid Y])^2\big]

For each value of $Y$, since $X \mid Y=y$ is Bernoulli with parameter $p$, we use $p(1-p)$.

For $Y=0$,

$$
\mathrm{Var}(X \mid Y=0)

Now we take expectation over $Y$:

$$
\mathbb{E}[\mathrm{Var}(X \mid Y)]

Substituting values,

$$
\mathbb{E}[\mathrm{Var}(X \mid Y)]
= \frac{5}{36}\cdot 0.60 + \frac{1}{4}\cdot 0.40.
$$

Rewriting probabilities,

$$
= \frac{5}{36}\cdot \frac{3}{5} + \frac{1}{4}\cdot \frac{2}{5}

= \frac{11}{60}.

Hence,

$$
\text{MSE}_{\min} = \frac{11}{60} \approx 0.1833.
$$

So we compare:

$$
\text{MSE}_{\min} = 0.1833,
\qquad
\text{MSE}_{\text{global}} = 0.21.
$$

Hence, using $Y$ reduces the error.

---

Using the Law of Total Variance:

$$
\mathrm{Var}(X) = \mathbb{E}[\mathrm{Var}(X \mid Y)] + \mathrm{Var}(\mathbb{E}[X \mid Y]).

Here,

$$
\mathbb{E}[\mathrm{Var}(X \mid Y)] = 0.1833,
$$

and

$$
\mathrm{Var}(\mathbb{E}[X \mid Y]) = 0.0267.
$$

Therefore,

$$
0.21 = 0.1833 + 0.0267.
$$

This shows that the reduction in error comes from the variability explained by $Y$.

$$
H(X\mid Y=0)

$$

$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.

# Conditional Density of the Multivariate Gaussian

Let

$$
Z=
\begin{bmatrix}
X\\
Y
\end{bmatrix}
\sim \mathcal{N}(\mu,\Sigma)
$$

with

$$
\mu=
\begin{bmatrix}
\mu_X\\
\mu_Y
\end{bmatrix},
\qquad
\Sigma=
\begin{bmatrix}
\Sigma_{XX} & \Sigma_{XY}\\
\Sigma_{YX} & \Sigma_{YY}
\end{bmatrix}.
$$

The joint density of $(X,Y)$ is

$$
f_{X,Y}(x,y)=
\frac{1}
{(2\pi)^{(n+m)/2}|\Sigma|^{1/2}}
\exp\left(
-\frac12
\begin{bmatrix}
x-\mu_X\\
y-\mu_Y
\end{bmatrix}^{T}
\Sigma^{-1}
\begin{bmatrix}
x-\mu_X\\
y-\mu_Y
\end{bmatrix}
\right).
$$

We want to derive the conditional density $f_{X|Y}(x|y)$.

## Step 1: Definition of Conditional Density

By definition,

$$
f_{X|Y}(x|y)
=
\frac{f_{X,Y}(x,y)}{f_Y(y)}.
$$

Since $Y$ is Gaussian,

$$
f_Y(y)=
\frac{1}
{(2\pi)^{m/2}|\Sigma_{YY}|^{1/2}}
\exp\left(
-\frac12
(y-\mu_Y)^T
\Sigma_{YY}^{-1}
(y-\mu_Y)
\right).
$$

Therefore,

$$
f_{X|Y}(x|y)
=
\frac{
|\Sigma_{YY}|^{1/2}
}{
(2\pi)^{n/2}|\Sigma|^{1/2}
}
\exp\left[
-\frac12(Q-R)
\right],
$$

where

$$
Q=
\begin{bmatrix}
x-\mu_X\\
y-\mu_Y
\end{bmatrix}^{T}
\Sigma^{-1}
\begin{bmatrix}
x-\mu_X\\
y-\mu_Y
\end{bmatrix},
$$

and

$$
R=
(y-\mu_Y)^T
\Sigma_{YY}^{-1}
(y-\mu_Y).
$$

## Step 2: Schur Complement

Define the Schur complement

$$
\Sigma_{X|Y}
=
\Sigma_{XX}
-
\Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}.
$$

Using the block matrix inverse identity,

$$
\Sigma^{-1}
=
\begin{bmatrix}
\Sigma_{X|Y}^{-1}
&
-\Sigma_{X|Y}^{-1}\Sigma_{XY}\Sigma_{YY}^{-1}
\\
-\Sigma_{YY}^{-1}\Sigma_{YX}\Sigma_{X|Y}^{-1}
&
\Sigma_{YY}^{-1}
+
\Sigma_{YY}^{-1}
\Sigma_{YX}
\Sigma_{X|Y}^{-1}
\Sigma_{XY}
\Sigma_{YY}^{-1}
\end{bmatrix}.
$$

Substituting into $Q$ and simplifying gives

$$
Q-R
=
(x-\mu^*)^T
\Sigma_{X|Y}^{-1}
(x-\mu^*),
$$

where

$$
\mu^*
=
\mu_X
+
\Sigma_{XY}\Sigma_{YY}^{-1}(y-\mu_Y).
$$

## Step 3: Determinant Identity

Using the determinant identity for block matrices,

$$
|\Sigma|
=
|\Sigma_{YY}|
|\Sigma_{X|Y}|.
$$

Hence,

$$
\frac{
|\Sigma_{YY}|^{1/2}
}{
|\Sigma|^{1/2}
}
=
\frac1
{|\Sigma_{X|Y}|^{1/2}}.
$$

## Step 4: Final Conditional Density

Substituting everything back,

$$
f_{X|Y}(x|y)
=
\frac{
1
}{
(2\pi)^{n/2}
|\Sigma_{X|Y}|^{1/2}
}
\exp\left(
-\frac12
(x-\mu^*)^T
\Sigma_{X|Y}^{-1}
(x-\mu^*)
\right).
$$

Therefore,

$$
X|Y=y
\sim
\mathcal{N}
\left(
\mu_X+\Sigma_{XY}\Sigma_{YY}^{-1}(y-\mu_Y),
\,
\Sigma_{XX}-\Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}
\right).
$$

Thus,

$$
\boxed{
\mu^*
=
\mu_X+\Sigma_{XY}\Sigma_{YY}^{-1}(y-\mu_Y)
}
$$

and

$$
\boxed{
\Sigma_{X|Y}
=
\Sigma_{XX}-\Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}
}
$$

which proves the required result.

The factory manager says: *"Since $Y = X + Q$, our best estimate of temperature should just be $Y$ itself."*
Using your result from Question 3, explain why $Y$ is a **biased** estimator of $X$ and why $E[X \mid Y]$ "shrinks" the measurement toward the mean (zero).



\textbf{1. Joint Distribution of $(X, Y)$:}

Since $X$ and $Q$ are independent normal random variables, $Y = X + Q$ is also normal.

\begin{align*}
E[Y] &= E[X] + E[Q] = 0 \\
\mathrm{Var}(Y) &= \mathrm{Var}(X) + \mathrm{Var}(Q) = 1 + 4 = 5 \\
\mathrm{Cov}(X,Y) &= E[X(X+Q)] = E[X^2] + E[XQ] = 1 + 0 = 1
\end{align*}

\[
\begin{pmatrix} X \\ Y \end{pmatrix}
\sim \mathcal{N}\!\left(
\begin{pmatrix} 0 \\ 0 \end{pmatrix},
\begin{pmatrix} 1 & 1 \\ 1 & 5 \end{pmatrix}
\right)
\]

\textbf{2. Information Relationship:}

$X$ is not $\sigma(Y)$-measurable since $Y$ does not uniquely determine $X$.

\textbf{3. Conditional Expectation:}

\[
E[X \mid Y=y]
= \mu_X + \frac{\mathrm{Cov}(X,Y)}{\mathrm{Var}(Y)}(y-\mu_Y)
= \frac{1}{5}y
\]

\[
E[X \mid Y] = 0.2Y
\]

\textbf{4. Residual Check:}

Let $Z = X - 0.2Y$.

\begin{align*}
E[Z] &= 0 \\
\mathrm{Cov}(Z,Y) &= E[(X-0.2Y)Y] \\
&= E[XY] - 0.2E[Y^2] \\
&= 1 - 0.2(5) = 0
\end{align*}

\textbf{5. Differential Entropy:}

\[
h(X) = \frac{1}{2}\ln(2\pi e)
\]

\[
\sigma^2_{X|Y} = 1 - \frac{1}{5} = 0.8
\]

\[
h(X|Y) = \frac{1}{2}\ln(2\pi e \cdot 0.8)
\]

\textbf{6. Mutual Information:}

\[
I(X;Y)
= h(X) - h(X|Y)
= \frac{1}{2}\ln\left(\frac{1}{0.8}\right)
= \frac{1}{2}\ln(1.25)
\]

\textbf{7. Error Comparison:}

\begin{align*}
\text{Naive estimator } (X=Y): &\quad \mathrm{Var} = 4 \\
\text{Optimal estimator } (E[X|Y]): &\quad \mathrm{Var} = 0.8
\end{align*}

The optimal estimator significantly reduces error, showing that $Y$ contains useful but noisy information about $X$.

```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from scipy.stats import multivariate_normal

# -----------------------------
# 1. Model parameters
# -----------------------------
mu_x, mu_q = 0, 0

var_x = 1
var_q = 4

# Y = X + Q
mu_y = mu_x + mu_q
var_y = var_x + var_q

cov_xy = var_x  # Cov(X, X+Q) = Var(X)
k = cov_xy / var_y  # shrinkage factor = 1/5

# Covariance matrix of (X, Y)
Sigma = np.array([
    [var_x, cov_xy],
    [cov_xy, var_y]
])

mu = np.array([mu_x, mu_y])

# -----------------------------
# 2. Generate samples
# -----------------------------
np.random.seed(42)
n = 1500

x = np.random.normal(mu_x, np.sqrt(var_x), n)
q = np.random.normal(mu_q, np.sqrt(var_q), n)
y = x + q

x_hat = k * y
residual = x - x_hat

# -----------------------------
# 3. Density grid
# -----------------------------
xg = np.linspace(-4, 4, 200)
yg = np.linspace(-8, 8, 200)
Xg, Yg = np.meshgrid(xg, yg)
pos = np.dstack((Xg, Yg))

Z = multivariate_normal(mu, Sigma).pdf(pos)

# -----------------------------
# 4. Figure
# -----------------------------
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=(
        "Joint Distribution (X, Y)",
        "Residual vs Y (Orthogonality)"
    )
)

# ---- Left: Joint distribution ----
fig.add_trace(
    go.Contour(
        x=xg, y=yg, z=Z,
        colorscale="Blues",
        opacity=0.5,
        showscale=False
    ),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(
        x=x, y=y,
        mode="markers",
        marker=dict(size=3, color="rgba(0,0,0,0.25)"),
        name="Samples (X, Y)"
    ),
    row=1, col=1
)

# Optimal estimator line: E[X|Y] = 0.2Y
y_line = np.linspace(-8, 8, 200)
fig.add_trace(
    go.Scatter(
        x=k * y_line,
        y=y_line,
        line=dict(color="red", width=3),
        name="E[X|Y] = 0.2Y"
    ),
    row=1, col=1
)

# Naive estimator X = Y
fig.add_trace(
    go.Scatter(
        x=y_line,
        y=y_line,
        line=dict(color="black", dash="dash"),
        name="Naive X = Y"
    ),
    row=1, col=1
)

# ---- Right: residual plot ----
fig.add_trace(
    go.Scatter(
        x=y,
        y=residual,
        mode="markers",
        marker=dict(size=3, color="purple", opacity=0.4),
        name="Residual"
    ),
    row=1, col=2
)

fig.add_trace(
    go.Scatter(
        x=y_line,
        y=np.zeros_like(y_line),
        line=dict(color="black"),
        name="Zero line"
    ),
    row=1, col=2
)

# -----------------------------
# 5. Layout fixes
# -----------------------------
fig.update_layout(
    title="Gaussian Signal + Noise Model (Shrinkage Estimation)",
    template="plotly_white",
    height=600
)

fig.update_xaxes(title_text="X", row=1, col=1)
fig.update_yaxes(title_text="Y", row=1, col=1)

fig.update_xaxes(title_text="Y", row=1, col=2)
fig.update_yaxes(title_text="Residual (X - 0.2Y)", row=1, col=2)

# keeps correct geometric scaling
fig.update_yaxes(scaleanchor="x", scaleratio=1)

fig.show()
```

**1. Shrinkage Principle (Conservative Estimation)**

In the left plot, we compare the naive estimator $X = Y$ with the optimal estimator $\mathbb{E}[X|Y] = 0.2Y$.

*   Since the noise (Torque $Q$, $\sigma^2 = 4$) is much larger than the signal (Temperature $X$, $\sigma^2 = 1$), the observed signal $Y$ is heavily noise-dominated.
*   The optimal estimator applies shrinkage: it reduces the magnitude of the observation by 80%.
*   For example, if $Y = 5$, then statistically only $1$ unit is attributed to $X$, while the remaining $4$ units are attributed to noise $Q$.
*   This prevents false alarms caused by random spikes in the observation.

**2. Orthogonality Principle (Optimality of Estimation)**

The right plot shows the residual error $Z = X - 0.2Y$ against the observation $Y$.

*   The scatter plot shows no visible correlation (horizontal structure).
*   This indicates that $Y$ contains no remaining predictive information about the error.
*   Mathematically, this is expressed as $\mathrm{Cov}(Z, Y) = 0$.
*   Hence, the estimator achieves the Minimum Mean Square Error (MMSE).

**3. Information Geometry and Measurability**

The contour plot illustrates why $X$ is not $\sigma(Y)$-measurable.

*   For any fixed value of $Y$, there is still variability in $X$, corresponding to conditional variance $0.8$.
*   The mapping from $(X,Q)$ to $Y$ reduces a 2D system into a 1D observation, causing information loss.
*   The conditional distribution represents irreducible uncertainty that cannot be eliminated using $Y$ alone.

**4. Variance Reduction and Mutual Information**

*   Prior uncertainty: $\mathrm{Var}(X) = 1.0$
*   Posterior uncertainty: $\mathrm{Var}(X|Y) = 0.8$
*   Mutual information: $I(X;Y) = 0.1116$ nats

This shows that observing $Y$ reduces uncertainty by 20%. To achieve a higher information gain, a different sensor less correlated with $Q$ would be required.
